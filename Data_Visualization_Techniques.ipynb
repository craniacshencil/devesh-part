**Data Visualization Techniques using matplotlib and seaborn**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv(r'C:\Users\dvesh\OneDrive\Desktop\AIML CLUB\Session 2\medical_insurance.csv')

# Display the first few rows of the dataset
print(df.head())
```

**Barcharts:**
A visual representation of data using rectangular bars of varying lengths to show the quantities or values associated with different categories.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate average charges by region
avg_charges_by_region = df.groupby('region')['charges'].mean()

# Plotting the bar chart
plt.figure(figsize=(10, 6))
avg_charges_by_region.plot(kind='bar', color='skyblue')
plt.title('Average Medical Insurance Charges by Region')
plt.xlabel('Region')
plt.ylabel('Average Charges')
plt.xticks(rotation=0)
plt.show()
```

**Line plots:** A graph that displays data points connected by straight lines, commonly used to depict trends or patterns over a continuous interval.

```python
# Calculate average charges by age
avg_charges_by_age = df.groupby('age')['charges'].mean()

# Plotting the line chart
plt.figure(figsize=(10, 6))
avg_charges_by_age.plot(kind='line', color='blue', marker='o')
plt.title('Average Medical Insurance Charges by Age')
plt.xlabel('Age')
plt.ylabel('Average Charges')
plt.grid(True)
plt.show()
```

**Scatter plots:** A graphical representation of individual data points on a two-dimensional plane, used to explore relationships between two continuous variables.

```python
# Scatter plot with color-coding by smoker status
plt.figure(figsize=(10, 6))
plt.scatter(df[df['smoker'] == 'yes']['bmi'], df[df['smoker'] == 'yes']['charges'], color='red', alpha=0.6, label='Smoker')
plt.scatter(df[df['smoker'] == 'no']['bmi'], df[df['smoker'] == 'no']['charges'], color='blue', alpha=0.6, label='Non-Smoker')
plt.title('Medical Insurance Charges vs. BMI (Colored by Smoker Status)')
plt.xlabel('BMI')
plt.ylabel('Charges')
plt.legend()
plt.grid(True)
plt.show()
```

**Histograms:** A graphical representation of the distribution of a dataset, depicting the frequency of data points within specified intervals, typically represented as bars.

```python


# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['bmi'], bins=20, color='purple', edgecolor='black', alpha=0.7)
plt.title('Distribution of BMI')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

**Pie charts:** A circular statistical graphic divided into slices to illustrate numerical proportions, with each slice representing a relative portion of the whole.

```python
# Calculate the count of individuals by gender
gender_counts = df['sex'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightskyblue'])
plt.title('Distribution of Insured Individuals by Gender')
plt.show()
```

**Heatmaps:** A graphical representation of data where values in a matrix are represented as colors, often used to visualize relationships or patterns in large datasets.

```python
# Calculate the correlation matrix
corr_matrix = df.corr()

# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
