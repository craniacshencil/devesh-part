## 1.7 Concatenation
- Adding rows to the dataset
- Adding new columns to the dataset
- Merging datasets

```python
# Add an extra record at the bottom of the table
extra_flat = pd.DataFrame({"city": ["Vadodara"], "area": [1000], "floor": [2], "price": [4000000]})
df = pd.concat([df, extra_flat], axis =0, ignore_index = True)
df
```

```python
# Add a new column "rooms" containing the rooms for each flat
bedrooms = pd.DataFrame({"rooms": [2, 3, 1, 3, 2, 3, 1, 3, 2, 3, 1, 2, 1, 3, 1, 1]})
df = pd.concat([df, bedrooms], axis = 1)
df
```

## 4.4 Understanding metrics
- What are metrics?
- Why do we need metrics?
- What goes into selecting a metric?

```python
# calculating mean absolute error on predictions and actual values
from sklearn.metrics import mean_absolute_error
from numpy import mean
predicted = [3, 4, 5, 6]
actual    = [2.9, 4.2, 5.5, 5.8]
mae = mean_absolute_error(actual, predicted)
maep = mae * 100 / mean(actual)
print("The mean absolute error is: ", round(mae, 2))
print("The mean absolute error percentage is: ", round(maep, 2), "%")
```

image.png

```python
# calculate precision, recall, accuracy_score
from sklearn.metrics import accuracy_score, precision_score, recall_score
predicted = [0, 0, 1, 1, 1, 1]
actual    = [1, 1, 0, 1, 1, 1]
print("The accuracy of our model is: ", accuracy_score(actual, predicted))
print("The recall of our model is: ", recall_score(actual, predicted))
print("The precision of our model is: ", precision_score(actual, predicted))
```
